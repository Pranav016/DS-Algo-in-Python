# Binary Search

```python
import numpy as np
```

```python
np.random.seed(0)
```

```python
def binary(a, key, start, end):
    if start>=end:
        return
    mid=(start+end)//2
    if a[mid]==key:
        return mid
    elif a[mid] > key:
        return binary(a, key, start, mid)
    else:
        return binary(a, key, mid+1, end)
```

```python
x=0
arr= np.random.randint(1000, size=100)
x=binary(arr, arr[50], 0, 100)
if x:
    print("Element found at {}".format(x))
else:
    print("Element not found")
```
